```python
def Hour_description(df,Columns,variables_str):
    df_new= pd.DataFrame()
    df_new_Columns=[]
    Hourini=6 # Hora de inicio
    Hourfini=14 # Hora final
    figx=4
    figy=3
    fig, axs = plt.subplots(figx, figy, tight_layout=True,figsize=(10,12))
    bins=20
    count=Hourini
    for i in range(figx):
        for j in range(figy):

            d=df.loc[df.HOUR==count,Columns]
            axs[i,j].hist(d, bins=bins)
    
            # Get statistics
            min_val = d.min()[0]
            max_val = d.max()[0]
            mean_val = d.mean()[0]
            med_val = d.median()[0]
    
            axs[i,j].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2,label='Hola')
            axs[i,j].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)
            axs[i,j].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
            axs[i,j].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    
            axs[i,j].grid()
            axs[i,j].set_title(f'Histogram of {variables_str} at {count} hours')
    
            df_new_Columns=['Hour_{}'.format(count)]
            df_new=df_new.append(d, ignore_index=True)
            df_new.rename(columns = {str(Columns[0]):df_new_Columns[0]}, inplace = True)
            count=count+1

            
    return df_new
```

```python
Columns,variables_str=['VOLTAGE_AVERAGE_(V)'],'Voltages'
df_VOLTAGES=Hour_description(df,Columns,variables_str)
```
